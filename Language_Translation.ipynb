## 📝 Step 2: Write the App Code

```python
app_code = '''
import streamlit as st
from googletrans import Translator
from gtts import gTTS
import base64
import io
import time

st.set_page_config(page_title="LinguaAI — Language Translation", page_icon="🌐", layout="centered")

st.markdown("""
<style>
@import url(\'https://fonts.googleapis.com/css2?family=Inter:wght@400;600;700&family=Space+Grotesk:wght@700&display=swap\');
html,body,[class*="css"]{font-family:\'Inter\',sans-serif;}
.stApp{background:linear-gradient(135deg,#0f0c29,#302b63,#24243e);}
#MainMenu,footer,header{visibility:hidden;}
.hero{text-align:center;padding:2rem 1rem 1.5rem;}
.hero h1{font-family:\'Space Grotesk\',sans-serif;font-size:2.4rem;color:#fff;}
.hero h1 span{background:linear-gradient(90deg,#a78bfa,#60a5fa);-webkit-background-clip:text;-webkit-text-fill-color:transparent;}
.hero p{color:#94a3b8;}
.card{background:rgba(255,255,255,0.05);border:1px solid rgba(255,255,255,0.1);border-radius:14px;padding:1.6rem;margin-bottom:1rem;}
.out-box{background:rgba(167,139,250,0.08);border:1px solid rgba(167,139,250,0.3);border-radius:10px;padding:1rem 1.2rem;color:#e2e8f0;font-size:1.05rem;line-height:1.7;}
.badge{display:inline-block;background:rgba(96,165,250,0.15);color:#60a5fa;border-radius:6px;padding:.15rem .6rem;font-size:.72rem;font-weight:600;letter-spacing:.5px;margin-bottom:.6rem;}
</style>
""", unsafe_allow_html=True)

LANGUAGES = {
    "Auto Detect":"auto","English":"en","Hindi":"hi","French":"fr","Spanish":"es",
    "German":"de","Chinese (Simplified)":"zh-cn","Arabic":"ar","Japanese":"ja",
    "Korean":"ko","Portuguese":"pt","Russian":"ru","Italian":"it","Dutch":"nl",
    "Turkish":"tr","Bengali":"bn","Punjabi":"pa","Tamil":"ta","Telugu":"te",
    "Urdu":"ur","Gujarati":"gu","Marathi":"mr","Thai":"th","Vietnamese":"vi",
}
LANG_NAMES = {v:k for k,v in LANGUAGES.items()}
translator = Translator()

st.markdown(\'\'\'
<div class="hero">
  <div class="badge">AI POWERED · 20+ LANGUAGES</div>
  <h1>Lingua<span>AI</span></h1>
  <p>Instant translation with Text-to-Speech</p>
</div>\'\'\', unsafe_allow_html=True)

st.markdown(\'<div class="card">\', unsafe_allow_html=True)
c1,c2 = st.columns(2)
with c1:
    src = st.selectbox("Source Language", list(LANGUAGES.keys()), index=0)
with c2:
    tgt_opts = [k for k in LANGUAGES if k!="Auto Detect"]
    tgt = st.selectbox("Target Language", tgt_opts, index=tgt_opts.index("Hindi"))
st.markdown(\'</div>\', unsafe_allow_html=True)

st.markdown(\'<div class="card">\', unsafe_allow_html=True)
text = st.text_area("", placeholder="Type or paste your text here...", height=130, label_visibility="collapsed")
st.caption(f"📝 {len(text)} chars · {len(text.split()) if text.strip() else 0} words")
st.markdown(\'</div>\', unsafe_allow_html=True)

_,btn_col,_ = st.columns([1,2,1])
with btn_col:
    go = st.button("🔄 Translate Now", use_container_width=True, type="primary")

if go:
    if not text.strip():
        st.warning("⚠️ Please enter some text.")
    else:
        with st.spinner("Translating..."):
            try:
                t0 = time.time()
                res = translator.translate(text, src=LANGUAGES[src], dest=LANGUAGES[tgt])
                ms = round((time.time()-t0)*1000)
                detected = LANG_NAMES.get(res.src, res.src.upper())
                st.markdown(\'<div class="card">\', unsafe_allow_html=True)
                st.markdown("**📄 Translation**")
                st.markdown(f\'<div class="out-box">{res.text}</div>\', unsafe_allow_html=True)
                st.caption(f"Detected: {detected} → {tgt} · {ms}ms")
                st.code(res.text)
                try:
                    tts = gTTS(res.text, lang=LANGUAGES[tgt] if LANGUAGES[tgt]!="auto" else "en")
                    buf = io.BytesIO()
                    tts.write_to_fp(buf); buf.seek(0)
                    b64 = base64.b64encode(buf.read()).decode()
                    st.markdown(f\'<audio controls style="width:100%"><source src="data:audio/mp3;base64,{b64}" type="audio/mp3"></audio>\', unsafe_allow_html=True)
                    st.caption("🔊 Listen to translation")
                except: st.info("Audio unavailable for this language.")
                st.markdown(\'</div>\', unsafe_allow_html=True)
                st.success("✅ Translation done!")
            except Exception as e:
                st.error(f"❌ Error: {e}")

st.markdown("<br><center><small>Built by Navya · CodeAlpha AI Internship · Task 1</small></center>", unsafe_allow_html=True)
'''

with open('app.py', 'w') as f:
    f.write(app_code)

print('✅ app.py written successfully!')
```

## 🧪 Step 5: Test Translation (Without UI)

```python
from googletrans import Translator

translator = Translator()

test_cases = [
    ("Hello, how are you?", "hi"),
    ("Artificial Intelligence is the future.", "fr"),
    ("I love machine learning.", "es"),
    ("Good morning!", "ja"),
]

print('🧪 Translation Test Results')
print('='*60)
for text, target in test_cases:
    result = translator.translate(text, dest=target)
    print(f'📝 Input  ({result.src.upper()}): {text}')
    print(f'✅ Output ({target.upper()}): {result.text}')
    print('-'*40)
```

## ⏹️ Step 6: Stop the App

```python
ngrok.disconnect(public_url)
proc.terminate()
print('🛑 App stopped.')
```
